//LOC (localização) usa dois parâmetros: linhas e colunas

> Bloco com recuo



```python
import pandas as pd
```

```python
df_prouni = pd.read_excel('/content/cursos-prouni.csv.xlsx')
df_prouni
```

```python
df_teste = df_prouni[['grau', 'turno', 'curso_busca', 'uf_busca', 'cidade_busca']]
df_teste
```

```python
df1 = df_teste.loc[:101,'curso_busca']
df1
```

//Trazendo mais colunas com o LOC: Passe uma lista com as colunas

```python
df2 = df_teste.loc[:101,['curso_busca', 'uf_busca', 'cidade_busca']]
df2
```

//Enquanto o LOC trabalha com o nome da coluna, o comando ILOC trabalha com a posição da coluna

```python
df3 = df_prouni.iloc[:100, 1]
df3
```

```python
df4 = df_prouni.iloc[:100,:]
df4
```

```python
df5 = df_prouni.iloc[:100, 2:4]
df5
```

//Função Reset Index

```python
df_prouni.reset_index()
df_prouni
```

//Esta função é muito útil qnd se trabalha com filtro, pois o filtro deixará o indice de linhas bagunçado. Retonará uma nova coluna chamada index. Para que isso não aconteça, utilizar o parametro (drop = True)

```python
df_prouni.reset_index(drop = True)
df_prouni
```

//O reset gera uma cópia do original. Caso queira alterar o df original, é necessário utilizar o parâmetro: (inplace = True)

```python
df_prouni.reset_index(inplace= True ,drop = True)
df_prouni
```

```python
empresa = {'email': ['m.antonia@gmail.com', 'c.soares@gmail.com'],
           'departamento': ['marketing', 'ti'],
           'nome': ['Maria Antônia', 'Cristina Soares']}

df_empresa = pd.DataFrame(empresa)
df_empresa
```

//Alterando o índice:

```python
df_empresa.set_index('email')
```

// A mudança sempre é feita em uma cópia. Para Alterar definitivamente é necessário utilizar o parâmetro *inplace*

```python
df_empresa.set_index('email', inplace= True)
df_empresa
```

//Utilizando Append para juntar duas tabelas

```python
df_clientes = pd.read_excel('/content/CLIENTES.xlsx')
df_produtos = pd.read_excel('/content/PRODUTO.xlsx')
```

```python
df_client_append = df_clientes.append(df_produtos)
df_client_append
```

//O comando append ira juntar as tabelas, uma embaixo da outra, porém, mantendo o índice inicial de cada tabela. Para que o índice fique em ordem, podemos utilizar o comando ignore_index

```python
df_client_append = df_clientes.append(df_produtos, ignore_index= True)
df_client_append
```

```python
df_cliente_produto = pd.concat([df_clientes, df_produtos])
df_cliente_produto
```
